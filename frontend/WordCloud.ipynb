```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from PIL import Image
from nltk.stem.porter import *

import matplotlib.pyplot as plt

# Concatenando todas as críticas em uma única variável
all_reviews = " ".join(s for s in df_classificado['Review'])
```

```python
print('Quantidade de Palavras: {}'.format(len(all_reviews)))
```

```python
# lista de stopword
stopwords = set(STOPWORDS)
stopwords.update(['e','dia', 'que', 'da', 'meu', 'você', 'de', 'ao', 'o','os', 'para'])
# Gerar um wordcloud
wordCloud = WordCloud(stopwords = stopwords, background_color='white', width=1600, height=800).generate(all_reviews)

# mostrar a imagem final
fig, ax = plt.subplots(figsize=(10,6))
ax.imshow(wordCloud, interpolation='bilinear')
ax.set_axis_off()

plt.imshow(wordCloud);
wordCloud.to_file("Dados\wordcloud_reviews.png")
```
