```python
import pandas as pd
import glob
```

```python
%time
path = r'C:\Users\efvs\Documents\Residencia Ednael\Dados despesas'
filenames = glob.glob(path + "/*.csv")

li= []

for filename in filenames:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

df_all = pd.concat(li, axis=0, ignore_index=True)
```

```python
df_all.info()
```

```python
df_all=df_all.astype({'Nome Favorecido':'category',
                        'Sigla UF':'category',
                        'Nome Município':'category',
                        'Código Órgão Superior':'int',
                        'Nome Órgão Superior':'category',
                        'Código Órgão':'int',
                        'Nome Órgão':'category',
                        'Código Unidade Gestora':'int',
                        'Nome Unidade Gestora':'category'})
```
