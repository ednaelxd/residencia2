```python
#varre todos o objetos e identifica se possui relação com obras
#caso sim, nomeia
def eh_obra(x):
    y=str()
    if ('OBRAS' in x) or ('OBRA' in x) or ('ENGENHARIA' in x) or ('engenharia' in x) or ('obras' in x) or ('obra' in x) or ('CONTRU-CAO' in x) or ('constru-cao' in x):
        y='OBRAS'
    return(y)
```

```python
#aplica a função de descobrimento de obras de forma mais rápida no dataframe
def separaObra(df2):
    return pd.Series(eh_obra(row.Descrição)
        for row in df2.itertuples()
  )
```

```python
abc=pd.DataFrame(separaObra(lic_item))
```

```python
def categorize(x):
    y=[]
    if x<=10000:
        y='Gasto_T1'
    elif x>10000 and x<=500000:
        y='Gasto_T2'
    elif x>500000 and x<=1000000:
        y='Gasto_T3'
    elif x>1000000 and x<=5000000:
        y='Gasto_T4'
    elif x>5000000 and x<=10000000:
        y='Gasto_T5'
    elif x>10000000:
        y='Gasto_T6'
    return(y)
```

```python
df = lic_item.head()
```

```python
df.iloc[0,-4]
```

```python
def itertuples_impl(df2):
    return pd.Series(categorize(row.total)
        for row in df2.itertuples()
  )
```

```python
itertuples_impl(lic_item)
```
