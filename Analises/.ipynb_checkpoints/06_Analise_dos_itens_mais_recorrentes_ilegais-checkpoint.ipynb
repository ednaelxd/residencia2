```python
# Carregar o Dataset de Licitacao completo
df_licitacao_2018_a_2021 = pd.read_csv(r'D:\itemLicitacao_Licitacao\dados_taguiados.csv')
```

### Agora faremos a junção do dataset de Itens Licitação e Licitação  utilizando o Número  do Processo como atributo de correlação entre as duas bases de dados.

```python
m = pd.merge(df_licitacao_2018_a_2021, df_item_licitacao_2018_a_2021, how = 'inner', on = 'Número Processo')
```

```python
m
```

```python
m.shape
```

```python
m.info()
```

```python
# Criar um novo DataFrame para selecionar apenas as freatures relevantes para a análise
df_licitacao_itens = m.loc[:, ['Número Licitação_x', 'Modalidade Compra_x', 'Número Processo', 'data', 'Total', 'Tipo', 'Tag', 'Descrição','Quantidade Item', 'Valor Item']]
```

```python
df_licitacao_itens
```

```python
# Renomeando algumas freatures para um melhor entendimento
df_licitacao_itens.rename(columns={'Número Licitação_': 'Num. Licitação', 'Modalidade Compra_x': 'Modalidade Compra', 'Total': 'Total Lic.','Quantidade Item': 'Quant','Valor Item': 'Total Itens',}, inplace=True)
```

```python
# Adicionando uma nova freature para extrair os valores de cada item.
df_licitacao_itens['Valor Item'] = df_licitacao_itens['Total Itens'] / df_licitacao_itens['Quant']
```

```python
df_licitacao_itens.head()
```

```python
df_licitacao_itens['Tag'].value_counts()
```

### Separando a Base de dados em Licitações Legais(1) e Ilegais(0)

#### **1. Licitações Legais**

```python
# Criando um objeto para separar as bases de dados de acordo com as tags
obj = df_licitacao_itens.groupby('Tag')
```

```python
# Separando os grupos do objeto 
obj.groups
```

```python
# Coletando as licitações Legais
df_legais = obj.get_group(1)
```

```python
df_legais.head()
```

```python
df_legais.shape
```

```python
df_legais['Tag'].value_counts()
```

## Visão Geral dos valores de cada item por modalidade de Preço na Licitações Legais

```python
descricao_valor_itens_legais = df_legais.groupby(['Descrição', 'Modalidade Compra']).agg({'Valor Item':['sum','mean','std','max','min', 'count']})
```

```python
descricao_valor_itens_legais = descricao_valor_itens_legais['Valor Item']
```

```python
descricao_valor_itens_legais.head(20)
```

```python
df_legais.loc[:,['Descrição', 'Modalidade Compra', 'Quant']].value_counts().head(20)
```

#### **2.  Licitações Ilegais**

```python
# Coletando apenas as licitações classificadas como ilegais
df_ilegais = obj.get_group(0)
```

```python
df_ilegais.head()
```

```python
df_ilegais.shape
```

```python
df_ilegais['Tag'].value_counts()
```

## Visão Geral dos valores de cada item por modalidade de Preço na Licitações Ilegais

```python
descricao_itens_ilegais = df_ilegais.groupby(['Descrição', 'Modalidade Compra']).agg({'Valor Item':['sum','mean','std','max','min', 'count']})
```

```python
descricao_itens_ilegais = descricao_itens_ilegais['Valor Item']
```
