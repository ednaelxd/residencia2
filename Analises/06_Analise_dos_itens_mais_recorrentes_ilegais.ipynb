### Agora faremos a junção do dataset de Itens Licitação e Licitação  utilizando o número do processo como atributo de correlação entre as duas bases de dados.

```python
m = pd.merge(df_licitacao_2018_a_2021, df_item_licitacao_2018_a_2021, how = 'inner', on = 'Número Processo')
```

```python
m
```

```python
m.shape
```

```python
m.info()
```

```python
# Criar um novo DataFrame para selecionar apenas as freatures relevantes para a análise
df_licitacao_itens = m.loc[:, ['Número Licitação_x', 'Número Processo', 'data', 'Total',  'Tipo', 'Tag', 'Descrição','Quantidade Item', 'Valor Item']]
```

```python
df_licitacao_itens
```

```python
# Renomeando algumas freatures para um melhor entendimento
df_licitacao_itens.rename(columns={'Total': 'Total Lic.','Quantidade Item': 'Quant','Valor Item': 'Total Itens',}, inplace=True)
```

```python
# Adicionando uma nova freature para extrair os valores de cada item.
df_licitacao_itens['Valor Item'] = df_licitacao_itens['Total Itens'] / df_licitacao_itens['Quant']
```

```python
df_licitacao_itens.head()
```

```python
# Coletando apenas as licitações classificadas como ilegais
df_licitacao_itens_ilegais = df_licitacao_itens[df_licitacao_itens['Tag'] == 0]
```

```python
df_licitacao_itens_ilegais
```

### Visualizando os itens mais recorrentes com seus respectivos valores e quantidades.

```python
df_licitacao_itens_ilegais.loc[:,['Descrição', 'Total Lic.', 'Quant', 'Valor Item', 'Total Itens']].value_counts().head(20)
```

```python
resumo_descricao_itens = df_licitacao_itens_ilegais.groupby(['Descrição']).agg({'Valor Item':['sum','mean','std','max','min']})
```
